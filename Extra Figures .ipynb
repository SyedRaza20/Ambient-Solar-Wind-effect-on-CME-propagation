## The uni_mae_plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file (replace with your file path if needed)
file_path = 'DONKI_CMEs_cleaner.csv'
df = pd.read_csv(file_path)

# Convert the 'time_21_5' column to datetime format
df['time_21_5'] = pd.to_datetime(df['time_21_5'])

# Extract the year from the 'time_21_5' column
df['year'] = df['time_21_5'].dt.year

# Count the number of CMEs per year
cme_count_per_year = df['year'].value_counts().sort_index()

# Ensure all years are included, even if the count is 0
cme_count_per_year_full = pd.Series(index=range(2012, 2024), data=0)
cme_count_per_year_full.update(cme_count_per_year)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(cme_count_per_year_full.index, cme_count_per_year_full.values, marker='o', linestyle='-', color='b')
plt.title('Number of CMEs per Year (2012-2023)', fontsize = 16)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of CMEs', fontsize = 14)
plt.xticks(cme_count_per_year_full.index)  # Show all years on the x-axis
plt.grid(True)

# Display the plot
plt.savefig("figures/nCMEs.png", dpi = 500)
plt.show()
```
